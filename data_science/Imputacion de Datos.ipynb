```python
#df_missing  = df_missing.drop(columns='ind')
to_replace = {'famhist':{'Present':1,
                         'Absent' :0} }

df_missing = df_missing.replace(to_replace)
df_missing = df_missing.drop(columns='ind')
df_missing
```

```python
from sklearn.impute import KNNImputer

imputer       = KNNImputer(n_neighbors=3)
df_impute_knn = pd.DataFrame(imputer.fit_transform(df_missing),columns=df_missing.columns)
df_impute_knn[query]
```

```python
pd.concat((df_complete.loc  [query,col],
           df_impute        [query]    ,
           df_impute_knn.loc[query,col]    
          ),axis=1)
```

```python
@interact
def impute_comparison(select_colum=['ldl','age'], k=(2,50,1)):
    query = df_missing[select_colum].isna() #missing data in c@lumns select_colum

    df_impute   = pd.DataFrame()
    mean_impute = df_missing.loc[~query,select_colum].mean()

    c_name                   = "{}_by_mean".format(select_colum)
    df_impute[c_name]        = df_missing[select_colum]
    df_impute[c_name][query] = mean_impute

    median_impute = np.median(df_missing.loc[~query,select_colum])

    c_name                   = "{}_by_median".format(select_colum)
    df_impute[c_name]        = df_missing[select_colum]
    df_impute[c_name][query] = median_impute

    imputer       = KNNImputer(n_neighbors=k)
    df_impute_knn = pd.DataFrame(imputer.fit_transform(df_missing),
                                   columns=df_missing.columns)

    return pd.concat((df_complete.loc  [query,select_colum],
                      df_impute        [query]    ,
                      df_impute_knn.loc[query,select_colum]    
                     ),axis=1)
```
